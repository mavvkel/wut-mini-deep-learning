```python
import os
import IPython.display as ipd
from scipy.io import wavfile
```

```python
org_f = './prep_dataset_no_resamp/train/left/00176480_nohash_0.wav'
resamp_f = './prep_dataset/train/left/00176480_nohash_0.wav'

org_sample_rate, org_samples = wavfile.read(os.path.join(os.getcwd(), org_f))
resamp_sample_rate, resamp_samples = wavfile.read(os.path.join(os.getcwd(), resamp_f))

print(f"Orginal len: {len(org_samples)}")
print(f"Resample len: {len(resamp_samples)}")

print(f"Orginal file size: {os.path.getsize(org_f)}")
print(f"Resample file size: {os.path.getsize(resamp_f)}")
```

```python
ipd.Audio(org_samples, rate=org_sample_rate)
```

```python
ipd.Audio(resamp_samples, rate=resamp_sample_rate)
```
